```python
import numpy as np
import csv

len_data = 10954

# Set the mean and variance
occ_areas = np.linspace(start=0, stop=1, num=101)
occ_areas = np.round(occ_areas, 4)

for area in occ_areas:

    # Generate uniformly distributed occlusion heights
    occ_heights = np.random.uniform(area, 1, (len_data, 1))
    
    # Calculate occlusion widths that lead to a consistent occluded area
    occ_widths = area / occ_heights

    # Concatenate widths and heights
    occ_shapes = np.concatenate((occ_heights, occ_widths), axis=1)

    
    # Generate random positions for the occlusions (inside the possible space)
    margin = np.random.uniform(0, 1 - occ_shapes)
    
    # Concatenate all
    data = np.concatenate((occ_shapes, margin), axis=1)

    # Round the data
    data = np.round(data, 4)

    # # Check the shape of the generated data
    # print("Data shape:", data.shape)
    # print(data)

    # Save the data to a CSV file
    filename = f'occlusions/uniformly_distributed/rect_occlusions_{area}.csv'
    with open(filename, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerows(data)

    print(f"CSV file '{filename}' created successfully.")
```
